```python
import xml.sax as sax
import re
import pandas as pd
```

```python
class qaHandler(sax.ContentHandler):
    
    def __init__(self,outFilename,types,xdelim):
        self.types = types
        self.filep = open(outFilename,"w")
        self.xdelim = xdelim
        self.curtype = None
        self.store = False
        self.curdict = {}
        for k in self.types:
            self.filep.write(k + ",")
            self.curdict[k] = ""
        
    def startElement(self,name,attr):
        
        if name in self.types:
            self.store = True
            self.curtype = name
            
    
    def endElement(self,name):
        if name == self.xdelim:
            self.filep.write("\n")
            
            for k in self.types:
                self.filep.write(self.curdict[k] + ",")
                self.curdict[k] = ""
            self.store = False
        if self.curtype == name:
            self.store = False
        
    def characters(self,content):
        if self.store:
            if not (content == "<" or content == "br /" or content == ">"):
                #print(content)
                self.curdict[self.curtype] += re.sub("\n|,"," ",content.strip())
    
    def endDocument(self):
        self.filep.close()
        
        
    
```

```python
xdelim = 'vespaadd'
storeTypes = ['subject','content','bestanswer','qlang']
```

```python
parser = sax.make_parser()
parser.setFeature(sax.handler.feature_namespaces, 0)
parser.setContentHandler(qaHandler("./LangMoreMinData.csv",storeTypes,xdelim))
```

```python
parser.parse("./data/FullOct2007.xml")
```
